```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# ข้อมูลจุดสำหรับสอง class
red_points = np.array([[1, 1], [2, 2], [2, 3]])
blue_points = np.array([[3, 1], [3, 2], [4, 3]])

# รวมข้อมูลทั้งหมดและสร้าง labels
X = np.vstack((red_points, blue_points))
y = np.array([0, 0, 0, 1, 1, 1])  # 0 = สีแดง, 1 = สีน้ำเงิน

# สร้าง SVM model โดยใช้ linear kernel
model = SVC(kernel='linear')
model.fit(X, y)

# ดึงพารามิเตอร์ของ SVM สำหรับ decision boundary
w = model.coef_[0]
b = model.intercept_[0]
x_values = np.linspace(0, 5, 100)
y_values = -(w[0] / w[1]) * x_values - b / w[1]

# Plot จุดข้อมูล
plt.figure(figsize=(8, 6))
plt.scatter(red_points[:, 0], red_points[:, 1], color='red', label='Red Class')
plt.scatter(blue_points[:, 0], blue_points[:, 1], color='blue', label='Blue Class')

# Plot decision boundary
plt.plot(x_values, y_values, color='black', label='Decision Boundary')

# Plot support vectors
support_vectors = model.support_vectors_
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], edgecolor='black', color='yellow', s=150, marker='o', label='Support Vectors')

# กำหนดรายละเอียดในกราฟ
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('SVM Decision Boundary with Support Vectors')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid()
plt.show()
```
