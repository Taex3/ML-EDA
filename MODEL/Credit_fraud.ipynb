```python
import math
num_features = len(df.columns[:-1])  
num_cols = 3  
num_rows = math.ceil(num_features / num_cols)  

plt.figure(figsize=(15, num_rows * 5))  
for i, col in enumerate(df.columns[:-1], 1):  
    plt.subplot(num_rows, num_cols, i)
    sns.boxplot(x=df["Class"], y=df[col])
    plt.title(f"Boxplot of {col} by Class")

plt.tight_layout()
plt.show()
```

```python
# คำนวณจำนวนฟีเจอร์และ grid
num_features = len(df.columns[:-1])  # ไม่รวม 'Class'
num_cols = 3  # กำหนดจำนวนคอลัมน์ของ subplot
num_rows = math.ceil(num_features / num_cols)  # คำนวณจำนวนแถวที่ต้องการ

plt.figure(figsize=(15, num_rows * 5))  # ปรับขนาดกราฟให้เหมาะสม
for i, col in enumerate(df.columns[:-1], 1):  # ไม่รวม 'Class'
    plt.subplot(num_rows, num_cols, i)  # ปรับ subplot ให้รองรับทุกฟีเจอร์
    skewness = df[col].skew()  # คำนวณ skewness
    sns.histplot(df[col], kde=True, label=f"Skewness: {skewness:.3f}")  # แสดง distribution
    plt.legend()
    plt.title(f"Distribution of {col}")

plt.tight_layout()  # จัด layout ให้ดูดี
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['Time'], bins=50, kde=True, color='blue')
plt.title('Distribution of Time')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()
```

```python
X=df.drop(columns='Class')
y=df['Class']
```

```python
class_counts = y.value_counts()
class_labels = ['Non-Fraud', 'Fraud']
plt.figure(figsize=(6, 4))
class_counts.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Class Distribution')
plt.xticks(ticks=[0, 1], labels=class_labels, rotation=0)
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (30, 30))
sns.heatmap(df.corr(), fmt = ".3f", annot = True, cmap = "crest")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=1234)
```

Dataset ไม่มีความ imbalance Class Distributions 'Non-Fraud', 'Fraud'มีจำนวนที่ต่างกันมากเลยต้องทำการSMOTE

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=1234)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report


knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train_resampled, y_train_resampled)
y_pred_knn = knn.predict(X_test)

print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
print("KNN Classification Report:\n", classification_report(y_test, y_pred_knn))
```

```python
from sklearn.tree import DecisionTreeClassifier


dt = DecisionTreeClassifier(max_depth=10, min_samples_split=5, criterion='gini', random_state=42)
dt.fit(X_train_resampled, y_train_resampled)
y_pred_dt = dt.predict(X_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Decision Tree Classification Report:\n", classification_report(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier


rf = RandomForestClassifier(n_estimators=100, max_depth=20, min_samples_split=5, random_state=42)
rf.fit(X_train_resampled, y_train_resampled)
y_pred_rf = rf.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train_resampled, y_train_resampled)
y_pred_nb = nb.predict(X_test)

print("Naive Bayes Accuracy:", accuracy_score(y_test, y_pred_nb))
print("Naive Bayes Classification Report:\n", classification_report(y_test, y_pred_nb))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=1, penalty='l2', solver='liblinear', random_state=42)
lr.fit(X_train_resampled, y_train_resampled)
y_pred_lr = lr.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_lr))
```

```python
new_data = X_test.iloc[:5]  
actual_values = y_test.iloc[:5].values  
```

```python
# พยากรณ์ด้วย Decision Tree
predictions_dt = dt.predict(new_data)
correct_dt = predictions_dt == actual_values
print("\nDecision Tree Predictions:", predictions_dt)
print("Actual Values:            ", actual_values)
print("Decision Tree Correct Predictions:", correct_dt)

# พยากรณ์ด้วย Random Forest
predictions_rf = rf.predict(new_data)
correct_rf = predictions_rf == actual_values
print("\nRandom Forest Predictions:", predictions_rf)
print("Actual Values:            ", actual_values)
print("Random Forest Correct Predictions:", correct_rf)

# พยากรณ์ด้วย Naive Bayes
predictions_nb = nb.predict(new_data)
correct_nb = predictions_nb == actual_values
print("\nNaive Bayes Predictions:", predictions_nb)
print("Actual Values:          ", actual_values)
print("Naive Bayes Correct Predictions:", correct_nb)

# พยากรณ์ด้วย Logistic Regression
predictions_lr = lr.predict(new_data)
correct_lr = predictions_lr == actual_values
print("\nLogistic Regression Predictions:", predictions_lr)
print("Actual Values:                ", actual_values)
print("Logistic Regression Correct Predictions:", correct_lr)
```
