```python
df_selected = merge_2[['Luminosity(L/Lo)','Radius(R/Ro)']]

from sklearn.ensemble import IsolationForest
iso_forest = IsolationForest(contamination=0.05, random_state=42)
iso_forest.fit(df_selected)

# Predict if each row is an outlier or not (1 means not an outlier, -1 means outlier)
outliers_if = iso_forest.predict(df_selected)
outliers_if = outliers_if == -1  # Convert prediction to True (outlier) or False (not an outlier)

#Filter out the latest outliers (current value -1)
cleaned_data = merge_2[outliers_if == 1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(merge_2)
scaled_df = pd.DataFrame(scaled, columns=merge_2.columns)
```

```python
#Split to X and Y
from sklearn.model_selection import train_test_split
target = ['Spectral Class']
X = scaled_df.drop(target, axis=1)
y = scaled_df[target]
```

```python
# 4. Split data into test and train (0.3 for test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_neighbors': list(range(1, 21)),
    'metric': ['euclidean', 'manhattan']
}
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
```

```python
print(f"Best parameters: {grid_search.best_params_}")
y_pred = grid_search.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
print(f"R2 Score: {r2_score(y_test, y_pred)}")
```

```python
# plot graph
results = pd.DataFrame(grid_search.cv_results_)

# Filter results for each metric
euclidean_results = results[results['param_metric'] == 'euclidean']
manhattan_results = results[results['param_metric'] == 'manhattan']

# Plot the results for each metric
plt.figure(figsize=(10, 6))
plt.plot(
    euclidean_results['param_n_neighbors'],
    euclidean_results['mean_test_score'],
    label='Euclidean',
    marker='o'
)
plt.plot(
    manhattan_results['param_n_neighbors'],
    manhattan_results['mean_test_score'],
    label='Manhattan',
    marker='o'
)

# Add details to the graph
plt.title('Grid Search Results')
plt.xlabel('Number of Neighbors (n_neighbors)')
plt.ylabel('Mean Accuracy')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
dt_model = DecisionTreeRegressor(random_state=1234)
```

```python
param_grid = {
    'criterion': ['mse', 'friedman_mse'],  
    'max_depth': list(range(1, 21)),
    'min_samples_split': list(range(2, 21)),  
    'min_samples_leaf': list(range(1, 21)),
}
```

```python
grid_search = GridSearchCV(estimator=dt_model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
```

```python
print(f"Best parameters: {grid_search.best_params_}")
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R Score: {r2}")
```

```python
# Visualize the decision tree
feature_names = X.columns
plt.figure(figsize=(45, 15))
plot_tree(
    best_model,
    filled=True,
    feature_names=feature_names,
    rounded=True,
    fontsize=12
)
plt.title("Decision Tree Regressor Visualization", fontsize=16)
plt.show()
```

```python
#  GridSearchCV  DataFrame
results = pd.DataFrame(grid_search.cv_results_)

# 
parameters_to_plot = ['param_max_depth', 'param_min_samples_split', 'param_min_samples_leaf']
colors = ['blue', 'green', 'red']
```

```python
#  plot 
plt.figure(figsize=(12, 8))

for param, color in zip(parameters_to_plot, colors):
    grouped_results = results.groupby(param)['mean_test_score'].mean()
    plt.plot(
        grouped_results.index,  #  x
        grouped_results.values,  #  y
        marker='o',
        label=param.replace('param_', '').replace('_', ' ').capitalize(),  #  label 
        color=color
    )

plt.title('Grid Search Results for Multiple Parameters', fontsize=16)
plt.xlabel('Parameter Value', fontsize=12)
plt.ylabel('Mean Negative MSE', fontsize=12)  #  y  MSE
plt.legend()
plt.grid()
plt.show()
```
