```python
import pandas as pd

# Load the data
file_path = "64011582_Potsatorn_Crosswalk_NIST and ISO27001.xlsx"
data = pd.ExcelFile(file_path)

# Parse the relevant sheet
sheet2_data = data.parse("Sheet2")

# Replace null values
sheet2_data['ISO/IEC 27001:2022'] = sheet2_data['ISO/IEC 27001:2022'].fillna('No Mapping Provided')

# Function to standardize ISO format
def standardize_iso_format(value):
    if "ISO/IEC 27001:2022" in str(value):
        clauses = value.split(",")
        return [clause.strip() for clause in clauses]
    return [value]

# Expand rows with multiple clauses
expanded_rows = []
for _, row in sheet2_data.iterrows():
    clauses = standardize_iso_format(row['ISO/IEC 27001:2022'])
    for clause in clauses:
        new_row = row.copy()
        new_row['ISO/IEC 27001:2022'] = clause
        expanded_rows.append(new_row)

# Create a cleaned DataFrame
cleaned_data = pd.DataFrame(expanded_rows)

# Save the cleaned data
cleaned_data.to_excel("cleaned_file.xlsx", index=False)
```
