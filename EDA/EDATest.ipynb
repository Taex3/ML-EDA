```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np



df_pets = pd.read_csv('pets_health_data.csv')
```

```python
df_pets.head()
```

```python
summary_stats = df_pets.describe()
summary_stats
```

```python
sns.pairplot(df_pets)
```

```python
sns.countplot(x='Species', hue='Health Issue (0=No, 1=Yes)', data=df_pets)
plt.title('Species vs Health Issue')
```

```python
sns.countplot(x='Vaccination', hue='Health Issue (0=No, 1=Yes)', data=df_pets)
plt.title('Vaccination vs Health Issue')
```

```python
sns.catplot(x='Vaccination', hue='Health Issue (0=No, 1=Yes)', col='Species',
            data=df_pets, kind='count', palette='Set1', hue_order=[0, 1])
```

```python
sns.boxplot(x='Health Issue (0=No, 1=Yes)', y='Age (Years)', data=df_pets)
plt.title('Age vs Health Issue')
```

```python
sns.boxplot(x='Health Issue (0=No, 1=Yes)', y='Weight (kg)', data=df_pets)
plt.title('Weight vs Health Issue')
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Weight (kg)', y='Health Issue (0=No, 1=Yes)', data=df_pets)

plt.title('Weight vs Health Issue')
plt.xlabel('Weight (kg)')
plt.ylabel('Health Issue (0=No, 1=Yes)')
plt.show()
```
