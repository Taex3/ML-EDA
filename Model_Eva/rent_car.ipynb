```python
import datetime as dt

class CarRental:
    def __init__(self, stock=0):
        self.stock = stock
    
    def stock(self):
        return self.stock
    
    def display_stock(self):
        print(f'we have currently {self.stock} cars available')
        return self.stock
    
    def rent_hourly(self, n):
        if n <= 0:
            print("the number of cars must be positive number")
            return None
        elif n > self.stock:
            print(f'Sorry, wehave only {self.stock} cars available')
            return None
        else:
            now = dt.datetime.now()
            print(f'you have rented {n} cars on hourly basis today at {now.hour}:{now.minute}:{now.second}')
            print('you will be charged 150 bath per hour')
            self.stock -= n
            return now
    
    def rent_daily(self, n):
        if n <= 0:
            print("the number of cars must be positive number")
            return None
        elif n > self.stock:
            print(f'Sorry, wehave only {self.stock} cars available')
            return None
        else:
            now = dt.date.today()
            print(f'you have rented {n} cars on dayly basis today on {now}')
            print('you will be charged 600 bath per day')
            self.stock -= n
            return now
        
    def rent_weekly(self, n):
        if n <= 0:
            print("the number of cars must be positive number")
            return None
        elif n > self.stock:
            print(f'Sorry, wehave only {self.stock} cars available')
            return None
        else:
            now = dt.date.today()
            print(f'you have rented {n} cars on dayly basis today on {now}')
            print('you will be charged 3000 bath per week')
            self.stock -= n
            return now
    
    def return_car(self, request):
        rental_time, rental_basis, num_cars = request
        bill = 0
        if rental_time and rental_basis and num_cars:
            self.stock += num_cars
            now = dt.datetime.now()
            rental_period = now - rental_time

            if rental_basis == 1:
                bill = round(rental_period.seconds / 3600) * 150 * num_cars
            elif rental_basis == 2:
                bill = round(rental_period.days) * 600 * num_cars
            elif rental_basis == 3:
                bill = round(rental_period.days / 7) * 3000 * num_cars
            
            if num_cars >= 3:
                print("you got discount ")
                bill = bill * 0.7
            print(f'The total bill is: {bill} bath')
        else:
            print('Are you sure you rent a cars ?')
            return None
```

```python
class Customer:
    def __init__(self):
        self.cars = 0
        self.rental_basis = 0
        self.rental_time = 0
        self.bill = 0
    
    def request_car(self):
        cars = input('how many car would you like to rent?')
        try:
            cars = int(cars)
        except ValueError:
            print('Invalid input: number must be interger')
            return -1
        if cars < 1:
            print('Invalid input: number must be positive numbers')
            return -1 
        else:
            self.cars = cars
        return self.cars
    
    def return_car(self):
        if self.rental_basis and self.rental_time and self.cars:
            return self.rental_time, self.rental_basis, self.cars
        else:
            return 0,0,0
```

```python
car = CarRental(10)
print('-------Hourly--------')
car.rent_hourly(3)
car.display_stock()

print('-------Daily--------')
car.rent_daily(3)
car.display_stock()

print('-------Weekly--------')
car.rent_weekly(3)
car.display_stock()
```

```python
cars_system = CarRental(10)
customer = Customer()

while True:
    print("""
    ====Car Rental System====
    1. Display available cars
    2.Request a car on houly basis - 150 bath
    3.Request a car on daliy basis - 600 bath
    4.Request a car on houly weekly - 3000 bath
    5. return cars
    6.Exit """)

    choice = input('Enter choice: ')
    try:
        choice = int(choice)
    except ValueError:
         print("the choice has to be from the available choices")
         continue
    if choice == 1:
        cars_system.display_stock()
    elif choice == 2:
        requested_car = customer.request_car()
        rental_time = cars_system.rent_hourly(requested_car)
        customer.rental_time = rental_time
        customer.rental_basis = 2
    elif choice == 3:
        customer.rental_time = cars_system.rent_daily(customer.request_car())
        customer.rental_basis = 3
    elif choice == 4:
        customer.rental_time = cars_system.rent_weekly(customer.request_car())
    elif choice == 5:
        request_tuple = customer.return_car()
        bill = cars_system.return_car(request_tuple)
        customer.bill = bill
        customer.rental_basis, customer.rental_time, customer.cars = 0,0,0
    elif choice ==6:
        break
    else:
        print('Invalid input')
print("THX for using the car rental system")
```
